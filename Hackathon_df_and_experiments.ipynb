### Category selection

```python
import pandas as pd
import re
import unicodedata
from typing import List, Tuple, Union

def _normalize(s: str) -> str:
    s = s.lower().strip()
    s = "".join(c for c in unicodedata.normalize("NFKD", s) if not unicodedata.combining(c))
    s = re.sub(r"[^\w\s\+\-\/\^]", " ", s)  # keep + - / ^
    s = re.sub(r"\s+", " ", s)
    return s

# Expanded blood patterns (Romanian + abbreviations + common variants)
blood_patterns = [
    # broad blood indicators
    r"\bsange\w*\b", r"\bsangelui\b", r"\bsingei\b", r"\bsingelui\b",

    # general blood panels
    r"\banaliza\s+generala\s+a\s+sangelui\b",
    r"\banaliza\s+partiala\s+a\s+sangelui\b",
    r"\bhemogram[aă]\b",
    r"\bhemoleucogram[aă]\b",
    r"\bformula\s+leucocitar[aă]\b", r"\bvsh\b",

    # biochemistry and hemostasis panels
    r"\bexamenul\s+biochimic\s+al\s+sangelui\b",
    r"\bexamenul\s+sistemului\s+de\s+hemostaza\b",
    r"\bcoagulograma\b",

    # coagulation
    r"\bindexul\s+protrombinic\b",
    r"\bactivitatea\s+protrombinei\b|\bquick\b",
    r"\binr\b",
    r"\btimpul\s+de\s+trombina\b|\btt\b",
    r"\btimpul\s+de\s+tromboplastina\s+partial[aă]?\s+activa\b|\b(aptt|ttpa)\b",
    r"\bcomplexe\s+fibrin[- ]monomer\b",
    r"\bfibrinogen\b",

    # inflammatory markers
    r"\bcrp\b|\bhs[- ]?crp\b|\bproteina\s+c[- ]reactiv[aă]?\b",
    r"\bpct\b|\bprocalcitonin[aă]?\b",
    r"\binterleukin[- ]?6\b",

    # enzymes & proteins
    r"\b(alat|alt)\b|\balanin[- ]?aminotransferaz[ae]\b",
    r"\b(asat|ast)\b|\baspart(al)?[- ]?aminotransferaz[ae]\b",
    r"\balbumin[ae]\b",
    r"\bggt\b|\bggtp\b|\bglutamiltranspeptidaz[ae]\b",
    r"\bck[- ]?mb\b|\bcreatinkinaz[ae](-| )?mb\b",
    r"\bck\b|\bcreatinkinaz[ae]\b",
    r"\bldh\b|\blactat\s+dehidrogenaz[ae]\b",
    r"\bamilaz[ae]\b",

    # ions / basic chemistry
    r"\bionograma\b",
    r"\bglucoz[ae](\s+din\s+deget)?\b|\bdozarea\s+glucozei(\s+in\s+sange)?\b",
    r"\bcreatinin[ae]\b",
    r"\buree\b",
    r"\b(calciu|ca)\b",
    r"\b(sodiu|na)\b",
    r"\b(potasiu|k)\b",
    r"\b(clor|cl)\b",
    r"\bproteina\s+total[aă]?\b",

    # bilirubin (all types)
    r"\bbilirubin[ae]\b",
    r"\bbilirubina\s+(directa|indirecta|totala)\b",

    # iron & ferritin
    r"\b(fier|fe)\b",
    r"\bferritin[ae]?\b",

    # D-dimer
    r"\bd[- ]?dimer(i)?\b",

    # BNP / cardiac markers
    r"\bbnp\b",
    r"\bnt[- ]?pro(bnp)?\b|\bctni\b|\bck[- ]?mb\b",

    # “dozarea” variants explicitly mentioning blood
    r"\bdozarea\s+amilazei\s+s(an)?g(e|a)i?\b",  # covers "amilazei singei"
]

blood_regex = re.compile("|".join(blood_patterns))

def classify_tests(input_data: Union[str, List[str]]) -> Tuple[List[object], List[object]]:
    """
    Classify input string(s) into blood tests and non-blood tests.
    Robust to diacritics/case/punctuation. Keeps NaNs in non-blood list.
    Returns (blood_list, non_blood_list) with original items preserved.
    """
    if isinstance(input_data, str):
        input_data = [input_data]

    blood_tests, non_blood_tests = [], []
    for item in input_data:
        if isinstance(item, str):
            norm = _normalize(item)
            if blood_regex.search(norm):
                blood_tests.append(item)
            else:
                non_blood_tests.append(item)
        else:
            non_blood_tests.append(item)  # keep NaNs/non-strings
    return blood_tests, non_blood_tests
```

```python
# Clean the data in order to extract the numbers from blood parameters

import pandas as pd
import numpy as np
import re

def convert_to_numeric_column(series):

    def clean_value(val):
        if pd.isna(val):
            return np.nan

        val = str(val).strip().lower()

        # Known invalid text patterns
        invalid_keywords = [
            'cheag', 'chiag', 'abs', 'blok', 'bloc', 'nsd', 'nu sunt teste',
            'lipsesc', 'nu se coag', 'nu se determ', '***', '****', '---', 'abs.'
        ]
        if any(x in val for x in invalid_keywords):
            return np.nan

        # Replace commas with dots for decimal conversion
        val = val.replace(",", ".")

        # Remove known units
        val = re.sub(r"\s*(mg/l|ng/ml|μg/ml|g/l|ml|mmol/l)", "", val)

        # Extract first numeric value
        match = re.search(r"[\d\.]+", val)
        if match:
            try:
                return float(match.group())
            except ValueError:
                return np.nan

        return np.nan

    # Apply cleaning function
    return series.apply(clean_value)


merge["clearned"] = convert_to_numeric_column(merge["value"])
merge
```
